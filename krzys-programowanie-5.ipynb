# Zadania

## Liczby

### Zadanie 1

Stwórz program w Pythonie który przyjmuje od użytkownika trzy liczby a następnie wypisuje na ekran informację która z tych liczb jest największa.

### Zadanie 2

Stwórz program w Pythonie który przyjmuje od użytkownika najpierw liczbę wartości, a następnie n kolejnych liczb. Program powinien potem wypisać największą z nich.

### Zadanie 3

Stwórz program w Pythonie który przyjmuje od użytkownika liczbę wartości, a następnie n kolejnych liczb. Program powinien zwrócić sumę tych liczb.

### Zadanie 4

Stwórz program w Pythonie który przyjmuje od użytkownika liczbę n, a następnie wypisuje kwadraty kolejnych liczb od 1 do n.

### Zadanie 5

Stwórz program w Pythonie który przyjmuje od użytkownika liczbę n, następnie kolejne n wartości, i wypisuje sumę **parzystych** liczb.

### Zadanie 6

Stwórz program w Pythonie który przyjmuje od użytkownika liczbę wartości, następnie kolejne n liczb, po czym wypisuje ile użytkownik wpisał wartości parzystych.

### Zadanie 7

Stwórz program w Pythonie który przyjmuje od użytkownika liczbę i wypisuje informację czy liczba jest czy nie jest pierwsza.

### Zadanie 8

Stwórz program w Pythonie który przyjmuje od użytkownika liczbę i wypisuje w kolejnych liniach wszystkie jej dzielniki.

### Zadanie 9

Stwórz program w Pythonie który przyjmuje od użytkownika liczbę i wypisuje sumę wszystkich jej dzielników.

### Zadanie 10

Stwórz program w Pythonie który przyjmuje od użytkownika liczbę i wypisuje na ekran największy możliwy kwadrat o boku równym liczbie całkowitej, którego pole jest mniejsze lub równe tej liczbie.

### Zadanie 11

Stwórz program który przyjmuje na wejściu trzy liczby a następnie informuje użytkownika czy z boków o podanych długościach można zbudować trójkąt.

### Zadanie 12

Stwórz program który przyjmuje na wejściu trzy liczby, po czym wypisuje je w kolejności od najmniejszej do największej. 

## Słowa

### Zadanie 1

Stwórz program w Pythonie który przyjmuje od użytkownika trzy słowa i wypisuje na ekran najdłuższe z nich.

### Zadanie 2

Stwórz program w Pythonie który przyjmuje od użytkownika trzy słowa i wypisuje na ekran ich połączenie.

### Zadanie 3

Stwórz program w Pythonie który przyjmuje od użytkownika słowo i zwraca je zapisane za pomocą wielkich liter.

### Zadanie 4

Stwórz program w Pythonie który przyjmuje od użytkownika zdanie (wiele słów) i zwraca je zapisane za pomocą wielkich liter.

### Zadanie 5

Stwórz program w Pythonie który przyjmuje od użytkownika słowo i liczbę, a następnie wypisuje na ekran to słowo powtórzone n razy.
