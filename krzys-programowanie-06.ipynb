# Zadania z programowania 6

## Zadanie 1

Stwórz program który przyjmie 5 liczb na wejściu i wypisze największą z nich.

## Zadanie 2

Stwórz program który przyjmie dowolną liczbę wartości na wejściu i wypisze największą z nich.

## Zadanie 3

Stwórz program który przyjmie 5 liczb na wejściu i poustawia je od najmniejszej do największej.

## Zadanie 4

Stwórz program który przyjmie dowolną liczbę wartości na wejściu i poustawia je od najmniejszej do największej.

## Zadanie 5

Stwórz program który przyjmie dowolną liczbę słów na wejściu i poustawia je w kolejności alfabetycznej.

## Zadanie 6

Stwórz program który przyjmuje na wejściu liczbę, a następnie wypisuje kolejne liczby z ciągu Collatza na ekran, po czym kończy gdy dojdzie do wartości 1.

## Zadanie 7

Stwórz program który przyjmie dowolną liczbę wartości na wejściu i poda różnicę pomiędzy najmniejszą a największą wartością.

## Zadanie 8

Stwórz program który przyjmie na wejściu dowolną liczbę wartości i wypisze ich średnią arytmetyczną.

## Zadanie 9 

Stwórz program który przyjmie na wejściu dowolną liczbę wartości i wypisze sumę ich kwadratów.

## Zadanie 10

Stwórz program który przyjmie na wejsciu dowolną liczbę wartosci i wypisze ich iloczyn.

## Zadanie 11

Stwórz program który przyjmie na wejściu dowolną liczbę wartości i wypisze ich średnią harmoniczną.

## Zadanie 12

Stwórz program który przyjmie na wejściu wartość `n` będącą wielkością obydwu list. Następnie przyjmie n wartości do pierwszej z list i n wartości do drugiej z list. Następnie powinien wypisać listę składającą się z sum wartości obydwu list.

## Zadanie 13

Stwórz program który przyjmie na wejściu liczbę `n` i słowo `x`, a następnie wypisze to słowo `n` razy w jednej linii. Przykład:

```
5
abc
```
powinno zwrócić `abcabcabcabcabc`

## Zadanie 14

Stwórz program który przyjmie na wejściu wartość `n` oraz listę `n` liczb 0 i 1, a następnie napisze ile razy każda z wartości się powtarza.

## Zadanie 15

Stwórz program który przyjmie na wejściu wartość `n` oraz listę `n` dowolnych liczb, a następnie wypisze które z tych liczb się powtarzają.
