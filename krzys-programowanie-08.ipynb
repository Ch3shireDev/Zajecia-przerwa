## Przypomnienie

Funkcje do manipulacji literami:

- `chr` zamienia liczbę na literę w alfabecie ascii. Zatem `chr(96)` jest równe `a`.
- `ord` zamienia literę na liczbę odpowiadającą pozycji w ascii. Zatem `ord('a')` jest równe 96.
- `input()` pozwala na pobranie wiadomości ze strumienia wejściowego
- `int()` pozwala na zamianę słowa na liczbę. Zatem `int('123')` zwróci `123`.
- `str()` pozwala na zamianę liczby na słowo. Zatem `str(123)` zwróci `'123'`.
- operacja `%` pozwala na zwrócenie reszty z dzielenia przez jakąś liczbę. Zatem `12 % 5` jest równe `2`.

## Zadanie 15

Stwórz program który przyjmie na wejście dwa słowa tej samej długości i wypisze ich "sumę" według ustalonej reguły - litery `a` do `z` mają wartości od `0` do `25`. `a` + `a`, czyli 0 + 0, powinno zwrócić 0, czyli `a`. `b` + `b`, czyli 1+1 powinno zwrócić 2, czyli `c`. Jeśli suma dwóch liczb przekroczyłaby 25, to brana jest wartość modulo - w ten sposób `o` + `o`, czyli 14 + 14 zwróci 28, czyli 2 modulo 26, zatem odpowiadająca tej kombinacji litera to `c`.

W ten sposób `abc` + `cde` powinno zwrócić `ceg`.
