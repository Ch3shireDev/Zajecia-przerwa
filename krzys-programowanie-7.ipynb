# Zadania z programowania

## Zadanie 1

Stwórz program który przyjmie słowo na wejściu i wypisze jego odwróconą wersję. Np. `abc` powinno zwrócić `cba`.

## Zadanie 2

Stwórz program który przyjmie na wejście dwie liczby i wypisze na ekran ich sumę.

## Zadanie 3

Stwórz program który przyjmie na wejście dwie liczby i wypisze na ekran ich różnicę.

## Zadanie 4

Stwórz program który przyjmie na wejście dwie liczby oraz nazwę operacji którą ma na nich wykonać:

- dodawanie
- odejmowanie
- mnożenie
- dzielenie

a następnie wypisze wynik wybranej operacji.

## Zadanie 5

Stwórz program który przyjmie literę i wypisze jej wartość ascii na ekran.

## Zadanie 6

Stwórz program który przyjmie wartość ascii i wypisze na ekran literę jej odpowiadającą.

## Zadanie 7

Stwórz program który przyjmie liczbę od 1 do 26 i wypisze n-tą literę alfabetu - `a` dla 1, `b` dla 2 itd.

## Zadanie 8

Stwórz program który przyjmie na wejście literę i wypisze odpowiadającą jej literę przesuniętą o 1 do przodu.


## Zadanie 9

Stwórz program który przyjmie na wejście literę i wypisze odpowiadającą jej literę przesuniętą o 1 do tyłu.

## Zadanie 10

Stwórz program który przyjmie na wejście literę oraz liczbę `n` i wypisze literę przesuniętą o `n` pozycji do przodu. Uwaga! `n` może być ujemne albo większe od 26.

## Zadanie 11

Stwórz program który przyjmie na wejście słowo i wypisze słowo z każdą literą przesuniętą o 1 do przodu.

## Zadanie 12

Stwórz program który przyjmie na wejście słowo i wypisze słowo z każdą literą przesuniętą o `n` liter do przodu.

## Zadanie 13

Stwórz program który przyjmie na wejście słowo i przesunie je "po okręgu" o jedną wartość - tj. `abcd` powinno zwrócić `dabc`, `kopytko` powinno zwrócić `okopytk` i tak dalej.

## Zadanie 14

Stwórz program który przyjmie na wejście słowo i liczbę `n` i przesunie słowo po okręgu `n`-krotnie.

## Zadanie 15

Stwórz program który przyjmie na wejście dwa słowa tej samej długości i wypisze ich "sumę" według ustalonej reguły - litery `a` do `z` mają wartości od `0` do `25`. `a` + `a`, czyli 0 + 0, powinno zwrócić 0, czyli `a`. `b` + `b`, czyli 1+1 powinno zwrócić 2, czyli `c`. Jeśli suma dwóch liczb przekroczyłaby 25, to brana jest wartość modulo - w ten sposób `o` + `o`, czyli 14 + 14 zwróci 28, czyli 2 modulo 26, zatem odpowiadająca tej kombinacji litera to `c`.

W ten sposób `abc` + `cde` powinno zwrócić `ceg`.
