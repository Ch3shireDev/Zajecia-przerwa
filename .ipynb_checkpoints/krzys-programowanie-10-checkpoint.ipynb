## Zadanie 10

Napisz podobny program co w zadaniu powyżej, tym razem jednak zliczając WSZYSTKIE litery. Podpowiedź: użyj struktury słownika.

```python
slownik = dict()
slownik['a'] = 0
print(slownik['a']) # zwraca 0
slownik['a'] += 1
print(slownik['a']) # zwraca 1
slownik['b'] = 3
slownik['c'] = 7

for key in slownik:
    print(key) #wypisze literę - hasło słownikowe, np. 'a'
    print(slownik[key]) # wypisze wartość pod hasłem słownikowym, dla 'a' będzie to 1
```

## Zadanie 20

Napisz program który pobierze zdanie i przetransformuje je w listę słów. Następnie każde słowo zostanie zamienione na parę składającą się z liczby określającej długość słowa i to słowo. Następnie tak uzyskana lista par ma zostać posortowana, i słowa mają zostać wypisane na ekran - od najkrótszego do najdłuższego.

Przykład:

```python
x = (1,'abc')
y = (3,'cde')
z = (2,'xyz')
tab = [x,y,z]
tab.sort()
for x,y in tab:
    print(y)
```

Wypisze pokolei:

```
abc
xyz
cde
```
