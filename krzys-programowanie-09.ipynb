# Zajęcia z programowania

## Zadanie 1

Stwórz program który przyjmuje liczbę na wyjściu i zwraca informację czy jest ona parzysta czy nie.

## Zadanie 2

Stwórz program który porównuje ze sobą dwie liczby i zwraca informację która z nich jest większa.

## Zadanie 3

Stwórz program z ustaloną wartością zmiennej. Użytkownik wywołując program proszony jest o podanie liczby, program następnie informuje czy liczba jest równa, mniejsza czy większa od danej zmiennej.

## Zadanie 4

Stwórz program identyczny do tego z zadania 3, z tą różnicą że zamiast przechowywać ustaloną wartość zmiennej program losuje ją na początku.

## Zadanie 5

Stwórz grę w 20 pytań - program losuje wartość zmiennej od 1 do 100 a następnie pyta o liczbę tak długo, dopóki użytkownik nie zgadnie wartości. Przy każdym podaniu wartości program odpowiada czy liczba jest za duża, za mała czy równa przechowywanej.

## Zadanie 6

Stwórz program który pyta użytkownika o literę i odpowiada czy występuje ona w słowie czy nie.

## Zadanie 7

Stwórz program który pyta użytkownika o litery tak długo aż użytkownik zgadnie wszystkie.

## Zadanie 8

Stwórz program który pyta użytkownika o wiek i wypisuje jego rok urodzenia.

## Zadanie 9

Stwórz program który pyta użytkownika o rok urodzenia i wypisuje jego wiek.

## Zadanie 10

Stwórz program który pyta użytkownika o datę urodzenia i wypisuje jego wiek.

## Zadanie 11

Stwórz program który pobiera od użytkownika czas w sekundach minutach i godzinach i konwertuje go na sekundy.

## Zadanie 12

Stwórz program który oblicza ile sekund minęło od danej daty.

## Zadanie 13

Zagadka matematyczna brzmi: ile świnek i kaczek ma rolnik, jeśli w sumie mają one 50 nóg i 15 głów? Stwórz program który zwróci informację na podstawie liczby świnek i liczby kaczek ile w sumie głów i nóg mają.

## Zadanie 14

Napisz program działający analogicznie jak w poprzednim zadaniu, jednak tym razem na podstawie nóg i głów ma zwracać on liczbę świń i kaczek.

## Zadanie 15

Napisz program konwertujący:

- a) funty na kilogramy
- b) kilogramy na funty
- c) celsjusze do fahreheitów
- d) fahrenheity do celsjuszów
- e) stopy na metry
- f) metry na stopy


## Zadanie 16

Napisz program obliczający na jaką wysokość poleci piłka rzucona z daną prędkością.
